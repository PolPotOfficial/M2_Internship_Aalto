```python
import geopandas as gpd
path2 = "/users/eleves-b/2023/wenrui.dai/Desktop/environment/Helsinki/h3_points_Helsinki.gpkg"
df2 = gpd.read_file(path2)
df2
```

```python
df2_copy = df2.copy()
```

```python
df2["id"] = df2["id"].astype(str)
df1["ORIGIN_CODE9"] = df1["ORIGIN_CODE9"].astype(str)
df1["DESTINATION_CODE9"] = df1["DESTINATION_CODE9"].astype(str)
```

```python
df2_indexed = df2.set_index("id")
df2_indexed
```

```python
origin_points = df2_indexed.loc[df1["ORIGIN_CODE9"]].reset_index()
destination_points = df2_indexed.loc[df1["DESTINATION_CODE9"]].reset_index()
```

```python
origins = gpd.GeoDataFrame(origin_points, geometry="geometry", crs=df2.crs)
destinations = gpd.GeoDataFrame(destination_points, geometry="geometry", crs=df2.crs)
```

```python
origins.to_file("TD_origins.gpkg", driver="GPKG")
destinations.to_file("TD_destinations.gpkg", driver="GPKG")
```
